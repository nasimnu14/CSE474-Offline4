```python
#Offline Task 2
import random 
import pandas as pd
import numpy as np
mean=0
classValue={"000":0,
         "001":1,
         "010":2,
         "011":3,
         "100":4,
         "101":5,
         "110":6,
         "111":7}
test_class={
    "11":1,
    "15":1,
    "21":1,
    "25":1,
    "32":1,
    "36":1,
    "42":1,
    "46":1,
    "53":1,
    "57":1,
    "67":1,
    "63":1,
    "78":1,
    "74":1,
    "88":1,
    "84":1       
}
def noise(mean=0,var=0.01):
    a=random.normalvariate(mean,var)
    return a

def channelOutput(h,Ik,Ik1):
    a=h[0]*Ik+h[1]*Ik1+noise()
    return a

def findmean(train_data):
    mean=np.zeros(shape=(8,2))
    for i in range(8):
        mean[i]=np.mean(train_data[train_data["class"]==i][["xk","xk1"]]).to_numpy()
    #print(mean)
    return mean
def testTransaction(i,j):
    st=""+str(i)+str(j)
    if st in test_class:
        return True
    else:
        return False
def formOutput(predicted):
    x=""
    size=len(predicted)
    for i in range(0,size):
        if(predicted[i] <=4):
            x+="0"
        else:
            x+="1"
    return x
def testFunction(y):
    size=len(y)
    d=np.zeros((y.size+1,8)) 
    parents=np.zeros((y.size+1,8))
    

    for k in range(1,size+1):
        if k==1:
            test_vec=y[0],0
        else:
            test_vec=y[k-1],y[k-2]
        for i in range(1,9):
            best_dist=float('inf')
            best_parent=0
            for j in range(1,9):
                if k==1 or testTransaction(j-1,i-1) ==True:
                    dist=d[k-1][j-1]+np.linalg.norm(test_vec-mean[i-1])
                if dist<best_dist:
                    best_dist=dist
                    best_parent=j
            d[k][i-1]=best_dist
            parents[k][i-1]=best_parent
    predicted=np.zeros(size)
    idx=int(np.argmin(d[size])+1)
    predicted[size-1]=idx
    for k in range(size,0,-1):
        
        par=parents[k][idx-1]
        #print(par)
        idx=int(par-1)
        predicted[k-1]=int(par)
    return formOutput(predicted)
    
        
    
if __name__ == '__main__':
    
    
    
    train=open("train.txt","r")
    train=train.read()
    parameter=open("parameter.txt","r")
    param=parameter.readline().split()
    mean=float(param[0])
    var=float(param[1])
    param=parameter.readline().split()
    h1=float(param[0])
    h2=float(param[1])
    h=[h1,h2]
    Ik1=int(train[0])
    Ik2=int(train[1])
    xk1=channelOutput(h,Ik1,Ik2)
    #print(xk1)
    l=len(train)
    train_data=np.empty(shape=(l-2,3))
    for i in range(2,l):
        Ik=int(train[i])
        class_w=str(Ik)+str(Ik1)+str(Ik2)
        class_w=classValue[class_w]
        
        
        xk=channelOutput(h,Ik,Ik1)
        
        train_data[i-2]=xk,xk1,class_w
        

        Ik2=Ik1
        Ik1=Ik
        xk1=xk
    train_data=pd.DataFrame(data=train_data,columns=["xk","xk1","class"])
    mean=findmean(train_data)
    test=open("test.txt","r")
    test=test.read()
    
    Ik1=0
    size=len(test)
    y=np.zeros(size)
    for i in range(0,size):
        Ik=int(test[i])
        y[i]= channelOutput(h,Ik,Ik1)
        Ik1=Ik
    output=testFunction(y)
    count=0
    for i in range(len(test)):
        if test[i]==output[i]:
            count+=1
    print("Accuracy is ",count)
```

```python
#Offline Task 1
import random 
import pandas as pd
import numpy as np
import scipy as sc
mean=0
pw=0
sigma=0
classValue={"000":0,
         "001":1,
         "010":2,
         "011":3,
         "100":4,
         "101":5,
         "110":6,
         "111":7}
test_class={
    "11":1,
    "15":1,
    "21":1,
    "25":1,
    "32":1,
    "36":1,
    "42":1,
    "46":1,
    "53":1,
    "57":1,
    "67":1,
    "63":1,
    "78":1,
    "74":1,
    "88":1,
    "84":1       
}
def noise(mean=0,var=0.01):
    a=random.normalvariate(mean,var)
    return a

def channelOutput(h,Ik,Ik1):
    a=h[0]*Ik+h[1]*Ik1+noise()
    return a

def findmean(train_data):
    mean=np.zeros(shape=(8,2))
    pw=[0]*8
    #print(train_data.shape[0])
    
    for i in range(8):
        mean[i]=np.mean(train_data[train_data["class"]==i][["xk","xk1"]]).to_numpy()
        pw[i]=train_data[train_data["class"]==i].shape[0]/train_data.shape[0]
    #print(mean)
    return mean,pw
def findSigma(train_data):
    sigma=np.zeros(shape=(8,2,2))
    #data=np.zeros(shape=(8,2))
    for i in range(0,8):
        data=train_data[train_data["class"]==i][["xk","xk1"]].to_numpy()
        cov=np.cov(data,rowvar=False)
        sigma[i,:,:]=cov
    return sigma
        
def testTransaction(i,j):
    st=""+str(i)+str(j)
    if st in test_class:
        return True
    else:
        return False
def formOutput(predicted):
    x=""
    size=len(predicted)
    for i in range(0,size):
        if(predicted[i] <=4):
            x+="0"
        else:
            x+="1"
    return x
def testFunction(y):
    size=len(y)
    d=np.zeros((y.size+1,8)) 
    parents=np.zeros((y.size+1,8))
    

    for k in range(1,size+1):
        if k==1:
            test_vec=y[0],0
        else:
            test_vec=y[k-1],y[k-2]
        for i in range(1,9):
            best_dist=float('inf')
            best_parent=0
            for j in range(1,9):
                if k==1 or testTransaction(j-1,i-1) ==True:
                    dist=d[k-1][j-1]+np.linalg.norm(test_vec-mean[i-1])
                if dist<best_dist:
                    best_dist=dist
                    best_parent=j
            d[k][i-1]=best_dist
            parents[k][i-1]=best_parent
    predicted=np.zeros(size)
    idx=int(np.argmin(d[size])+1)
    predicted[size-1]=idx
    for k in range(size,0,-1):
        
        par=parents[k][idx-1]
        #print(par)
        idx=int(par-1)
        predicted[k-1]=int(par)
    return formOutput(predicted)
    
        
    
if __name__ == '__main__':
    
    
    
    train=open("train.txt","r")
    train=train.read()
    parameter=open("parameter.txt","r")
    param=parameter.readline().split()
    mean=float(param[0])
    var=float(param[1])
    param=parameter.readline().split()
    h1=float(param[0])
    h2=float(param[1])
    h=[h1,h2]
    Ik1=int(train[0])
    Ik2=int(train[1])
    xk1=channelOutput(h,Ik1,Ik2)
    #print(xk1)
    l=len(train)
    train_data=np.empty(shape=(l-2,3))
    for i in range(2,l):
        Ik=int(train[i])
        class_w=str(Ik)+str(Ik1)+str(Ik2)
        class_w=classValue[class_w]
        
        
        xk=channelOutput(h,Ik,Ik1)
        
        train_data[i-2]=xk,xk1,class_w
        

        Ik2=Ik1
        Ik1=Ik
        xk1=xk
    train_data=pd.DataFrame(data=train_data,columns=["xk","xk1","class"])
    mean,pw=findmean(train_data)
    sigma=findSigma(train_data)
    print(sigma)
    test=open("test.txt","r")
    test=test.read()
    
    Ik1=0
    size=len(test)
    y=np.zeros(size)
    for i in range(0,size):
        Ik=int(test[i])
        y[i]= channelOutput(h,Ik,Ik1)
        Ik1=Ik
    output=testFunction(y)
    count=0
    for i in range(len(test)):
        if test[i]==output[i]:
            count+=1
    print("Accuracy is ",count)
```
